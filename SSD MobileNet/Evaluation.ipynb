## Evaluation
Let's do the evaluation process to see how well the chosen model will perform.<br>

It helps to measure the precision and recall where precision ensures that we are not misclassifying too many data whereas recall guarantees that we are not overlooking the data.<br>

*precision=truepositives/truepositives+falsepositives*<br>
*recall=truepositives/truepositives+falsenegatives*
